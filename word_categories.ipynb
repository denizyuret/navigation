```julia
using DataFrames

df = readtable("data/postags.csv");
```

```julia
nouns = Int[]
verbs = Int[]

for i=1:size(df,1)
    if startswith(df[i, 2], "NN")
        push!(nouns, i)
    elseif startswith(df[i, 2], "VB")
        push!(verbs, i)
    end
end

nns = df[nouns, :]
vbs = df[verbs, :]

sort!(nns, cols=:count, rev=true);
sort!(vbs, cols=:count, rev=true);
```
